```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv(r"heart_2020_cleaned.csv")
test = df.sample(10) #numero de valores a usar para prediccion
train = df[~df.isin(test)] #valores a usar para entrenamiento
train.dropna(inplace = True)
```

```python
train.head()
```

```python
from sklearn import preprocessing #convertimos los Yes y No de HeartDisease en binario
label = preprocessing.LabelEncoder()

train['HeartDisease']= label.fit_transform(train['HeartDisease'])
test['HeartDisease']= label.fit_transform(test['HeartDisease'])
print(train['HeartDisease'].unique())
print(test['HeartDisease'].unique())
```

```python
train.head()
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def square_loss(y_pred, target):
  return np.mean(pow((y_pred - target),2))
```

```python
X_tr, y_tr = train.BMI, train['HeartDisease'] #elegimos con que variable entrenamos y predecimos
X_te, y_te = test.BMI, test['HeartDisease']
```

```python
lr = 0.01 #learning late
W = np.random.uniform(0,1)
b = 0.1
for i in range(10000):
  z = np.dot(X_tr, W) + b
y_pred = sigmoid(z)
l = square_loss(y_pred, y_tr)
gradient_W = np.dot((y_pred-y_tr).T, X_tr)/X_tr.shape[0]
gradient_b = np.mean(y_pred-y_tr)
W = W - lr * gradient_W
b = b -lr* gradient_b
```

```python
for i in range(len(X_te)):
  r = sigmoid(np.dot(X_te, W) + b)
```

```python
print(r) # prediccion en base a BMI
```
